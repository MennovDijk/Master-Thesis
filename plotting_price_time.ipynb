```python
import pandas as pd
import numpy as np
from numpy import inf, nan
import datetime
from matplotlib import pyplot as plt
import statsmodels.formula.api as smf
import matplotlib.dates as mdates
import datetime
```

```python
def connectpoints(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot_date([x1,x2],[y1,y2],'r--', xdate = True, label = "trend")
```

```python
df = pd.read_csv("market_info_forge_daily.csv")
```

```python
column_name_plot = "cpqs"

count = 1
pltcount = 1
plt.figure(figsize=(60,60))

with open("item_ids_selected.txt") as f:
    while True:
        line = f.readline()
        if line == "":
            break
            
        else:
            n = int(line)
            df2 = df.loc[df["typeID"] == n]

            x = df2["date"]
            y = df2[column_name_plot]

            df2 = df2.assign(inc_dates = np.arange(len(x)))

            plt.subplot(30, 6, count)
            plt.plot_date(x, y, xdate = True, label = column_name_plot)

            plt.xlabel('Year')
            plt.ylabel('avgPrice')
            plt.legend(loc = 2)
            plt.title("typeID %s" % str(n))
            pltcount += 1
 
            count += 1

plt.savefig(column_name_plot + "_plot.png")
plt.show()
```
